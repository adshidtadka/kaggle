```python
test = pd.read_csv('../input/test.csv', parse_dates=["activation_date"])
train = pd.read_csv('../input/train.csv',  parse_dates=["activation_date"])
```

```python
test['is_train'] = False
train['is_train'] = True

all_df = pd.concat([test, train], axis = 0)
```

```python
all_df.tail()
```

```python
# New variable of weekday
all_df["activation_weekday"] = all_df["activation_date"].dt.weekday
```

```python
all_df.head()
```

```python
# New variable of holiday
all_df["activation_holiday"] = 0
for index, row in all_df.iterrows():
    if row[ "activation_weekday"] >= 6:
        all_df.at[index,  "activation_holiday"] = 1
```

```python
all_df.head()
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
```

```python
le.fit(all_df['item_id'])
```

```python
cat_vars = ["user_id", "region", "city", "parent_category_name", "category_name", "param_1", "param_2", "param_3", "user_type"]
for col in cat_vars:
    all_df[col] = all_df[col].astype('str')
    le.fit(all_df[col])
    all_df[col] = le.transform(all_df[col])
```

```python
cols_to_drop = ["item_id", "title", "description", "activation_date", "image"]
all_df = all_df.drop(cols_to_drop, axis = 1)
```
